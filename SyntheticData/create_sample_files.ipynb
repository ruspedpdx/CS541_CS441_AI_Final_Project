```python
from pydantic import BaseModel, Field, confloat
from typing import List, NamedTuple

class FileData(dspy.Signature):
    path = dspy.InputField(desc="The file, files, or directory")
    description = dspy.InputField(desc="The description of the file, files, or directory")
    file = dspy.OutputField(desc="No more than a 5 line sample of the files based on the description and path. Just include the file. Max 200 tokens")

GenerateSampleFile = dspy.Predict(FileData, temperature=0.6)
```

### Create more detailed file descriptions

```python
class MoreDetail(dspy.Signature):
    basic = dspy.InputField(desc="A basic description of the file")
    detailed = dspy.OutputField(desc="Describe the file in more detail. Imagine a specific use case. 20 words.")

generate_detailed_desc = dspy.Predict(MoreDetail, temperature=0.9)
```

```python
resp = generate_detailed_desc(basic=sample_files_dataset[10].Description)
print(resp)
```

```python
# List of file names
file_names = [f'./data/filetemp{i}.csv' for i in range(1, 6)]

# List to hold all entries
all_entries = []

# Read each file
for file_name in file_names:
    with open(file_name, 'r') as file:
        content = file.read()
        
        # Split the content by the specified punctuation
        entries = content.split('"\n"')
        
        # Extend the list with the split entries
        all_entries.extend(entries)

# Remove any empty strings from the list
all_entries = [entry for entry in all_entries if entry]

print(all_entries)
```

```python
all_entries = [entry.strip('"') for entry in all_entries]
all_entries = [entry.strip('`') for entry in all_entries]
all_entries = [entry.strip('"') for entry in all_entries]
```

```python
length_entries = len(all_entries)
print(length_entries)
print(all_entries[0])
print(all_entries[1])
print(all_entries[30])
```

```python
from collections import defaultdict

entry_indices = defaultdict(list)

# Populate the dictionary with indices
for index, entry in enumerate(all_entries):
    entry_indices[entry].append(index)

# Find duplicates
duplicates = {entry: indices for entry, indices in entry_indices.items() if len(indices) > 1}

# Print duplicates with their indices
if duplicates:
    print("Duplicates found:")
    for entry, indices in duplicates.items():
        print(f'"{entry}" occurs at indices {indices}')
else:
    print("No duplicates found")
```

```python
from collections import Counter

entry_counts = Counter(all_entries)
duplicates = {entry: count for entry, count in entry_counts.items() if count > 1}

# Print duplicates
if duplicates:
    print("Duplicates found:")
    for entry, count in duplicates.items():
        print(f'"{entry}" occurs {count} times')
else:
    print("No duplicates found")
```

### Make metric to choose the best one

```python
class MetricOutput(BaseModel):
    score: confloat(ge=1,le=5) = Field(...,description="Rate the sample of the file given the description")

class MetricSignature(dspy.Signature):
    description = dspy.InputField(desc="The description of what the file is")
    file = dspy.InputField(desc="The file to be judged")
    score: MetricOutput = dspy.OutputField()

get_score = dspy.TypedPredictor(MetricSignature)

#Non-standard implementation of metric
def rate_response(example, pred,trace=None):
    score_response = get_score(description=example.Description, file=pred.file)
    return score_response.score
```
